## 3. Final Correct Code

```python
class Solution:
    def moveZeroes(self, nums: list[int]) -> None:
        none_zeros = [num for num in nums if num != 0]
        zeros = len(nums) - len(none_zeros)

        nums[:] = none_zeros + [0] * zeros
```
